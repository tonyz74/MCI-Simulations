```python
import matplotlib.pyplot as plt
import arviz as az
import numpy as np

def plot_estimates(do_fn):
    from scipy.stats import norm

    values = [do_fn() for i in range(1000)]

    fig, ax = plt.subplots(figsize=(10, 6.2), layout="constrained")
    az.plot_dist(values, kind="hist", hist_kwargs={"bins": 30, "density": True}, ax=ax)
    
    ax.set_xlabel("Estimated Value of Integral")
    ax.set_ylabel("Relative Frequency")
    
    mu, stddev = norm.fit(values)
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100)
    p = norm.pdf(x, mu, stddev)
    plt.plot(x, p, 'k', linewidth=2)
    title = "Fit results: mu = %.6f,  std = %.10f" % (mu, stddev)
    plt.title(title)
    
    plt.locator_params(axis='x', nbins=8)
    plt.show()
```

```python
import vegas
import math
import random
import matplotlib.pyplot as plt
import arviz as az

def f(x):
    return math.exp(100 * -x ** 2)

plot_estimates(lambda: vegas.Integrator([-4, +4])(f, nitn=10, neval=100).mean);
```

```python
def naive(func, bounds, nsamp=1000):
    lo, hi = bounds

    sm = 0
    for i in range(nsamp):
        sm += func(random.random() * ((hi - lo)) + lo)
    return (sm * (hi - lo)) / nsamp

plot_estimates(lambda: naive(f, [-4, +4]))
```
